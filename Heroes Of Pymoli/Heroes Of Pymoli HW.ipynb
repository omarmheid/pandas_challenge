```python
purchase_df = pd.read_csv(file_to_load)

purchase_df.head()
```

```python
total_players = len(purchase_df["SN"].value_counts())
players_df = pd.DataFrame({"Total Players": total_players}, index = [0])

players_df
```

```python
unique_numbers = len(purchase_df["Item ID"].value_counts())
avg_purchase_price = purchase_df["Price"].mean()
total_purchases = len(purchase_df["Price"])
total_revenue = purchase_df["Price"].sum()


purchasing_analysis_df = pd.DataFrame({"Number of Unique Items": unique_numbers, 
                                      "Average Price": avg_purchase_price, 
                                      "Number of Purchases": total_purchases,
                                     "Total Revenue": total_revenue}, index = [0])

purchasing_analysis_df["Average Price"] = purchasing_analysis_df["Average Price"].map("${:.2f}".format)
purchasing_analysis_df["Total Revenue"] = purchasing_analysis_df["Total Revenue"].map("${:.2f}".format)

purchasing_analysis_df
```

```python
gender_grouped = purchase_df.groupby(["Gender"])
unique_members = gender_grouped["SN"].nunique()

gender_summary_df = pd.DataFrame({"Total Counts": unique_members, "Percentage of Players": 100*unique_members/total_players})
gender_summary_df.index.name = None
gender_summary_df["Percentage of Players"] = gender_summary_df["Percentage of Players"].map("{:.2f}%".format)
gender_summary_df
```

```python
purchase_count = gender_grouped["Purchase ID"].count()
average_purchase_price = gender_grouped["Price"].mean()
total_purchase = gender_grouped["Price"].sum()
average_purchase_total = total_purchase/unique_members

purchasing_analysis_gender_df = pd.DataFrame({"Purchase Count": purchase_count, 
                           "Average Purchase Price": average_purchase_price, 
                           "Total Purchase Value": total_purchase, 
                           "Average Total Purchase Per Person": average_purchase_total})
purchasing_analysis_gender_df.index.name = "Gender"


purchasing_analysis_gender_df["Average Purchase Price"] = purchasing_analysis_gender_df["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis_gender_df["Total Purchase Value"] = purchasing_analysis_gender_df["Total Purchase Value"].map("${:.2f}".format)
purchasing_analysis_gender_df["Average Total Purchase Per Person"]= purchasing_analysis_gender_df["Average Total Purchase Per Person"].map("${:.2f}".format)

purchasing_analysis_gender_df
```

```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]

group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_df["Age Group"] = pd.cut(purchase_df["Age"],age_bins, labels=group_names)
purchase_df

age_group = purchase_df.groupby("Age Group")

total_count_age = age_group["SN"].nunique()

percentage_by_age = (total_count_age/total_players) * 100

age_demographics_df = pd.DataFrame({"Total Count": total_count_age , "Percentage of Players": percentage_by_age})

age_demographics_df.index.name = None

age_demographics_df["Percentage of Players"] = age_demographics_df["Percentage of Players"].map("{:.2f}%".format)

age_demographics_df
                                                                                                
                                                                                


```

```python
purchase_count = age_group["Purchase ID"].count()
average_purchase_price = age_group["Price"].mean()
total_purchase = age_group["Price"].sum()
average_purchase_total = total_purchase/total_count_age

purchasing_analysis_age_df = pd.DataFrame({"Purchase Count": purchase_count,
                                       "Average Purchase Price": average_purchase_price,
                                       "Total Purchase Value": total_purchase, 
                                       "Average Total Purchase Per Person": average_purchase_total})

purchasing_analysis_age_df.index.name = "Age Ranges"



purchasing_analysis_age_df["Average Purchase Price"] = purchasing_analysis_age_df["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis_age_df["Total Purchase Value"] = purchasing_analysis_age_df["Total Purchase Value"].map("${:.2f}".format)
purchasing_analysis_age_df["Average Total Purchase Per Person"]= purchasing_analysis_age_df["Average Total Purchase Per Person"].map("${:.2f}".format)

purchasing_analysis_age_df

```

```python
top_spenders = purchase_df.groupby("SN")
purchase_count = top_spenders["Purchase ID"].count()
average_purchase_price = top_spenders["Price"].mean()
total_purchase_value = top_spenders["Price"].sum()

top_spenders_analysis_df = pd.DataFrame({"Purchase Count": purchase_count,
                                        "Average Purchase Price": average_purchase_price,
                                        "Total Purchase Value": total_purchase_value})

top_spenders_analysis_df["Average Purchase Price"] = top_spenders_analysis_df["Average Purchase Price"].map("${:.2f}".format)
top_spenders_analysis_df["Total Purchase Value"] = top_spenders_analysis_df["Total Purchase Value"].map("${:.2f}".format)

formatted_spenders_df = top_spenders_analysis_df.sort_values(["Total Purchase Value"], ascending=False).head()

formatted_spenders_df

```
